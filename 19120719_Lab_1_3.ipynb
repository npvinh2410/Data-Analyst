```python
import pandas as pd
import numpy as np
dict_columns={
    'numbers':['#23','#24','#18','#14','#12','#10','#35'],
    'nums':[23,24,18,14,np.nan,'XYZ',35],
    'colors':['green','red','yellow','orange','purple','blue','pink'],
    'other_column':[0,1,0,2,1,0,2]}
df = pd.DataFrame(dict_columns)
display(df)
print('Kieu du lieu tuong ung voi cac cot:')
df.dtypes
```

```python
#Cac xu ly gia tri cua 'numbers'
#Xoa ky tu '#' o cac gia tri cua 'numbers'
df=df.assign(numbers=df['numbers'].str.lstrip('#').apply(pd.to_numeric))
print('Gia tri mean cua numbers la: ')
df['numbers'].mean()
```

```python
#Ham in hoa chu cai dau
def inhoa(ch):
  ch1=''
  temp=''
  for i in range(0,len(ch)):
    if(i==0 or temp==' '):
      ch1+=ch[i].upper()
    else:
      ch1+=ch[i]
    temp=ch[i]
  return ch1
```

```python
#Chuyen cac gia tri Colors in hoa chu dau
df['colors']=df['colors'].apply(lambda x: inhoa(x))
display(df)
```

```python
#Kiem tra xem cac object co phai la so khong
def kiemtraso(obj):
  if(isinstance(obj,int)):
    #Neu la so thi tra ve chinh no
    return obj
  else:
    #Neu khong phai so tra ve NaN
    return np.nan
#Xu li data cot nums(chuyen cac doi tuong khong phai so ve NaN)
df['nums']=df['nums'].apply(lambda x: kiemtraso(x))
display(df)
df.dtypes
```

```python
#Xu ly missing data
#Thay the cac doi tuong NaN thanh gia tri mean cua cot nums
tb=int(df['nums'].mean())
df['nums'].fillna(tb,inplace=True)
display(df)
```
